```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize
from matplotlib.lines import Line2D

import mesa_reader as mr
```

```python
plt.style.use('./plotstyle.mplstyle')
```

## No TS Dynamo

```python
LOGS_FOLDER = './LOGS/'
```

```python
nts_h = mr.MesaData(LOGS_FOLDER+'history.data')
```

```python
h = nts_h
```

```python
cmap = cm.viridis

fig, ax = plt.subplots()

ax.scatter(h.log_Teff, h.log_L, c=h.center_he4, cmap=cmap)
ax.invert_xaxis()

ax.set_xlabel('$\\log T_\\mathrm{eff}$')
ax.set_ylabel('$\\log L/L_\\odot$')

cb = plt.colorbar(mappable=ScalarMappable(Normalize(vmin=0, vmax=1)), cmap=cmap, ax=ax)
cb_label = cb.set_label('Center He4')
```

```python
cmap = cm.viridis

fig, ax = plt.subplots()

ax.scatter(h.star_age, h.surf_avg_v_rot, c=h.center_he4, cmap=cmap)

ax.set_xlabel('$\\log T_\\mathrm{eff}$')
ax.set_ylabel('$v_\\mathrm{rot}\,(\\mathrm{km}\,\\mathrm{s}^{-1})$')

cb = plt.colorbar(mappable=ScalarMappable(Normalize(vmin=0, vmax=1)), cmap=cmap, ax=ax)
cb.set_label('Center He4')
```

```python
mld = mr.MesaLogDir(LOGS_FOLDER)

zams_profile = mld.profile_data(model_number=1)
tams_profile = mld.profile_data(model_number=157)
heex_profile = mld.profile_data()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(zams_profile.mass, zams_profile.omega, c='r', label='ZAMS')
ax.plot(tams_profile.mass, tams_profile.omega, c='b', label='TAMS')
ax.plot(heex_profile.mass, heex_profile.omega, c='g', label='HeExh')
ax.legend()

ax.set_xlabel('$M\,(M_\\odot)$')
ax.set_ylabel('$\\Omega\,(\\mathrm{s}^{-1})$')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(zams_profile.mass, zams_profile.log_j_rot, c='r', label='ZAMS')
ax.plot(tams_profile.mass, tams_profile.log_j_rot, c='b', label='TAMS')
ax.plot(heex_profile.mass, heex_profile.log_j_rot, c='g', label='HeExh')
ax.legend()

ax.set_xlabel('$M\,(M_\\odot)$')
ax.set_ylabel('$\\log j\,(M_\\odot^{2/3}\,\\mathrm{s}^{-1})$')
```

## TS dynamo


```python
LOGS_FOLDER = './LOGS_AM/'
```

```python
ts_h = mr.MesaData(LOGS_FOLDER+'history.data')
```

```python
h = ts_h
```

```python
cmap = cm.viridis

fig, ax = plt.subplots()

ax.scatter(h.log_Teff, h.log_L, c=h.center_he4, cmap=cmap)
ax.invert_xaxis()

ax.set_xlabel('$\\log T_\\mathrm{eff}$')
ax.set_ylabel('$\\log L/L_\\odot$')

cb = plt.colorbar(mappable=ScalarMappable(Normalize(vmin=0, vmax=1)), cmap=cmap, ax=ax)
cb.set_label('Center He4')
```

```python
cmap = cm.viridis

fig, ax = plt.subplots()

ax.scatter(h.star_age, h.surf_avg_v_rot, c=h.center_he4, cmap=cmap)

ax.set_xlabel('$\\log T_\\mathrm{eff}$')
ax.set_ylabel('$v_\\mathrm{rot}\,(\\mathrm{km}\,\\mathrm{s}^{-1})$')

cb = plt.colorbar(mappable=ScalarMappable(Normalize(vmin=0, vmax=1)), cmap=cmap, ax=ax)
cb.set_label('Center He4')
```

```python
mld = mr.MesaLogDir(LOGS_FOLDER)

zams_profile = mld.profile_data(model_number=1)
tams_profile = mld.profile_data(model_number=157)
heex_profile = mld.profile_data()
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(zams_profile.mass, zams_profile.omega, c='r', label='ZAMS')
ax.plot(tams_profile.mass, tams_profile.omega, c='b', label='TAMS')
ax.plot(heex_profile.mass, heex_profile.omega, c='g', label='HeExh')
ax.legend()

ax.set_xlabel('$M\,(M_\\odot)$')
ax.set_ylabel('$\\Omega\,(\\mathrm{s}^{-1})$')
ax.set_yscale('log')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(ts_h.star_age/1e6, ts_h.center_h1, 'k-', label='w/ TS')
ax.plot(nts_h.star_age/1e6, nts_h.center_h1, 'k--', label='w/o TS')

ax.plot(ts_h.star_age/1e6, ts_h.center_he4, 'b-', label='w/ TS')
ax.plot(nts_h.star_age/1e6, nts_h.center_he4, 'b--', label='w/o TS')

ax.plot(ts_h.star_age/1e6, ts_h.center_c12, 'g-', label='w/ TS')
ax.plot(nts_h.star_age/1e6, nts_h.center_c12, 'g--', label='w/o TS')

ax.plot(ts_h.star_age/1e6, ts_h.center_o16, 'r-', label='w/ TS')
ax.plot(nts_h.star_age/1e6, nts_h.center_o16, 'r--', label='w/o TS')

ax.set_xlabel('Age (Myr)')
ax.set_ylabel('Center')
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(ts_h.star_age/1e6, ts_h.center_c12, 'r-', label='w/ TS')
ax.plot(nts_h.star_age/1e6, nts_h.center_c12, 'b-', label='w/o TS')

ax.set_xlabel('Age (Myr)')
ax.set_ylabel('Center He4')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(zams_profile.mass, zams_profile.log_j_rot, c='r', label='ZAMS', lw=3)
ax.plot(tams_profile.mass, tams_profile.log_j_rot, c='b', label='TAMS', lw=3)
ax.plot(heex_profile.mass, heex_profile.log_j_rot, c='g', label='HeExh', lw=3)
ax.legend()

ax.set_xlabel('$M\,(M_\\odot)$')
ax.set_ylabel('$\\log j\,(M_\\odot^{2/3}\,\\mathrm{s}^{-1})$')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(zams_profile.mass, zams_profile.log_j_rot, 'r-', label='ZAMS', lw=3)
ax.plot(tams_profile.mass, tams_profile.log_j_rot, 'b-', label='TAMS', lw=3)
ax.plot(heex_profile.mass, heex_profile.log_j_rot, 'g-', label='HeExh', lw=3)

nts_mld = mr.MesaLogDir('LOGS')

nts_zams_profile = nts_mld.profile_data(model_number=1)
nts_tams_profile = nts_mld.profile_data(model_number=157)
nts_heex_profile = nts_mld.profile_data()

ax.plot(nts_zams_profile.mass, nts_zams_profile.log_j_rot, 'r--', lw=2)
ax.plot(nts_tams_profile.mass, nts_tams_profile.log_j_rot, 'b--', lw=2)
ax.plot(nts_heex_profile.mass, nts_heex_profile.log_j_rot, 'g--', lw=2)

handles, labels = ax.get_legend_handles_labels()
handles.append(Line2D((-2, -1), (0, 0), c='k', ls='-', label='w/ TS'))
handles.append(Line2D((-2, -1), (0, 0), c='k', ls='--', label='w/o TS'))
ax.legend(handles=handles, frameon=False)

ax.set_xlabel('$M\,(M_\\odot)$')
ax.set_ylabel('$\\log j\,(M_\\odot^{2/3}\,\\mathrm{s}^{-1})$')
ax.set_xlim(0, 15)
ax.set_ylim(13, 18.7)
```
